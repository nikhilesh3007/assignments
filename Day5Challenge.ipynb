Plotly offers interactive, scientific data visualization for web applications.
Plotly creates and stewards the leading open-source data visualization and app building libraries for ML, data science, engineering, and the sciences.
Their libraries support engineers and scientists publishing charts and analytics apps to the Web in Python, R, Julia, .NET, and JS.


It is used in web applications where there is a need of interactive, scientific data visualization.


SOURCE FOR THE CODE:- 
https://plotly.com/python/basic-charts/

```python
import plotly.graph_objects as go
#SOURCE= https://www.autocarindia.com/car-news/car-suv-sales-down-64-percent-in-may-2021-420998

labels = ['Maruthi_Suzuki_India','Hyundai_Motor_India','Tata_Motors','Kia_India','Mahindra&Mahindra','Renault_India','Honda_Cars_India','Volkswagen_India','Nissan_Motor_India','MG_Motor_India','Ford_india','Skoda_India','Toyota_Kirloskar_India','FCA','Citroen']
values = [32903, 25001, 15180, 11050, 8004, 2620, 2032, 1425, 1235, 1016, 766, 716, 707, 475, 40]

fig = go.Figure(data=[go.Pie(labels=labels, values=values,title="Auto Sales MAY 2021")])
fig.show()
```

```python
import plotly.graph_objects as go
#SOURCE= https://www.ndtv.com/coronavirus/india-covid-19-tracker
States=["Andhra Pradesh","Arunachal Pradesh","Assam","Bihar","Chhattisgarh","Goa","Gujarat","Haryana","Telangana","Tamilnadu","Kerala","Karnataka","Odisha","Maharashtra"]
Active_cases=[31850,3460,23048,1141,4914,1965,1497,1034,11206,33665,110616,38752,24737,117698]
fig = go.Figure([go.Bar(x=States, y=Active_cases)])
fig.update_layout(title_text="Covid-19 Active Cases in India")
fig.show()
```

```python
import plotly.express as px
#SOURCE= https://www.worldometers.info/coronavirus/country/india/
States=["Oct 2020","Nov 2020","Dec 2020","Jan 2021","Feb 2021","Mar 2021","Apr 2021","May 2021","Jun 2021","Jul 2021"]
Active_cases=[78745,48369,37161,19460,13829,16606,77199,319675,159399,44259]
fig = px.line(x=States, y=Active_cases, title="Covid-19 Recovery Rate in India")
fig.show()
```
